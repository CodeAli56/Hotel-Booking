# Importing Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

# Loading the Dataset

```python
data = pd.read_csv("data/hotel_booking.csv")
```

```python
data['reservation_status_date'] = pd.to_datetime(data['reservation_status_date'])
```

```python
data.drop(['agent', 'company'], axis=1, inplace=True)
data.dropna(inplace=True)
```

```python
data.describe().T
```

```python
data['adr'].plot(kind='box')
```

```python
data = data[(data['adr']>0) & (data['adr']<5000)]
```

```python
num_cols = data.select_dtypes(include=['number']).columns
data[num_cols].plot(kind='box', subplots=True, layout=(5, 4), figsize=(14, 8), sharex=False, sharey=False)
plt.tight_layout()
plt.show()
```

# Data Analysis And Visualization

```python
cancelled_perc = data['is_canceled'].value_counts(normalize=True)
print(cancelled_perc)
plt.figure(figsize=(5,4))
plt.bar(['Not Cancelled', 'Cancelled'], data['is_canceled'].value_counts(), width=0.5, edgecolor='k')
plt.title("Reservation Status Count")
plt.show()
```

```python
plt.figure(figsize=(7,4))
ax1 = sns.countplot(x='hotel', data=data, hue='is_canceled', palette='Blues')
plt.title("Reservation Status in different Hotels")
plt.xlabel("Hotels")
plt.ylabel("Reservation Count")
plt.show()
```

```python
resort_hotel = data[data['hotel'] == 'Resort Hotel']
resort_hotel['is_canceled'].value_counts(normalize=True)
```

```python
city_hotel = data[data['hotel']=='City Hotel']
city_hotel['is_canceled'].value_counts(normalize=True)
```

```python
resort_hotel_adr = resort_hotel.groupby("reservation_status_date")[['adr']].mean()
city_hotel_adr = city_hotel.groupby("reservation_status_date")[['adr']].mean()
```

```python
plt.figure(figsize=(20,5))
plt.plot(resort_hotel_adr, label='Resort Hotel')
plt.plot(city_hotel_adr, label='City Hotel')
plt.title("Average Daily Rate in City and Resort Hotel")
plt.legend(fontsize=10)
plt.show()
```

```python
data['month'] = data['reservation_status_date'].dt.month
plt.figure(figsize=(15,5))
ax1 = sns.countplot(x='month', data=data, hue='is_canceled')
plt.title("Reservation Status per Month")
plt.xlabel("Month Number")
plt.ylabel('Status Count')
plt.legend(['Not Cancelled',"Cancelled"])
plt.show()
```

```python
cancelled_data = data[data['is_canceled']==1]
```

```python
top_10_countries = cancelled_data['country'].value_counts(normalize=True)[0:10]

plt.figure(figsize=(6, 6))
plt.pie(top_10_countries, autopct="%.2f", labels=top_10_countries.index)
plt.title("Top 10 countries with reservation cancelled")
plt.show()
```

```python
reservation_sources = data['market_segment'].value_counts(normalize=True)[:5]
reservation_sources

plt.figure(figsize=(6, 6))
plt.pie(reservation_sources, autopct="%.2f", labels=reservation_sources.index)
plt.title("Top Reservation Channels")
plt.show()
```

```python
cancelled_sources = cancelled_data['market_segment'].value_counts(normalize=True)[:5]
cancelled_sources

plt.figure(figsize=(6, 6))
plt.pie(cancelled_sources, autopct="%.2f", labels=cancelled_sources.index)
plt.title("Top Cancelled Reservation Channels")
plt.show()
```

```python
cancelled_adr = cancelled_data[(cancelled_data['reservation_status_date'].dt.year < 2018) & (cancelled_data['reservation_status_date'].dt.year > 2015)].groupby('reservation_status_date')[['adr']].mean().reset_index().sort_values('reservation_status_date')
not_cancelled_adr = data[(data['is_canceled']==0) & (data['reservation_status_date'].dt.year < 2018) & (data['reservation_status_date'].dt.year > 2015)].groupby("reservation_status_date")[['adr']].mean().reset_index().sort_values('reservation_status_date')
```

```python
plt.figure(figsize = (20,6))
plt.plot(cancelled_adr['reservation_status_date'], cancelled_adr['adr'])
plt.plot(not_cancelled_adr['reservation_status_date'], not_cancelled_adr['adr'])
plt.legend(["Cancelled Adr",'Not Cancelled Adr'])
plt.title("Average Daily Rate (2016-2017)")
plt.show()
```
